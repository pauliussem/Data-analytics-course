```python
# Nuskaitome csv failus su Pandas

import pandas as pd

internet_usage = pd.read_csv('Final.csv')
regions = pd.read_csv('regions_csv.csv')
```

```python
# Susikūriau naujus stulpelius "Regionai", su reikšme 1 ir "Dešimtmetis", su reikšme 2.

internet_usage['Regionai'] = 1
internet_usage['Decade'] = 2
internet_usage
```

```python
# Regions dataFrame pasidarom į dictionary, kad prie šalių galėčiau prisiskirti regionus.

region_map = dict(zip(regions['Country'], regions['Region']))

for country in region_map:
    internet_usage.loc[internet_usage['Entity'] == country, 'Regionai'] = region_map[country]

internet_usage
```

```python
# Atsifiltravau informaciją, kur prie regionų įrašyti vienetukai.
# Faile regions_csv pakeičiau šalių pavadinimus, kad sutaptų su duomenimis iš failo Final.csv

internet_usage.loc[internet_usage['Regionai'] == "1"]
```

```python
# Prisiskiriam dešimtmečius.

def f(x):
  if x['Year'] >= 2000 and x['Year'] <= 2009: return '2000-2009'
  elif x['Year'] >= 2010 and x['Year'] <= 2019: return '2010-2019'
  elif x['Year'] >= 2020 and x['Year'] <= 2029: return '2020-2029'

internet_usage['Decade'] = internet_usage.apply(f, axis=1)

internet_usage
```

```python
# 1) Vidutinis interneto vartotojų skaičius "Cellular Subscription", "Internet Users(%)" ir "Broadband Subscription".

iu_avg = internet_usage[["Cellular Subscription", "Internet Users(%)", "Broadband Subscription"]].mean()

iu_avg
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.figure(figsize=(8,5))

labels = ['Cellular Subscription', 'Internet Users(%)', 'Broadband Subscription']
colors = ['#abcdef', '#aabbcc', '#eeefff']

plt.pie([75.873710, 32.186916, 8.487670], labels = labels, colors=colors, autopct='%.2f %%')

plt.title('Interneto vartojimo pasiskirstymas Pasaulyje, per paskutinius 20 metų.')

plt.show()
```

```python
# Susiskuriam atskira df su regionais, cellular, internet users ir broadband subscriptions.
# Susigrūpuojam regionus ir paskaičiuojam kiekvieno regiono vidurkius.

region_iu = internet_usage[['Regionai', 'Cellular Subscription', 'Internet Users(%)','Broadband Subscription']]

region_avg = region_iu.groupby(['Regionai'], as_index = False).mean()

region_avg


```

```python
# susikuriam regions setą ir susisortinam, kad visos šalys eitų iš eilės (kadangi duomenys df eina iš eilės)
# Į diagramą (x) susidedam regionų setus ir anksčiau apskaičiuotus regionų vidurkius (y).

import matplotlib.pyplot as plt
import numpy as np

region_avg.plot(figsize = (8, 5), kind='bar', x='Regionai', title = 'Interneto vartojimas pagal Žemynus', colormap = 'viridis')

plt.figure(figsize = (20,8))

plt.show()
```

```python
# Lygiai tą patį darome čia, pasiskaičiuojam vidurkius, tik dar išskyrdami šalis.

countries_by_region = internet_usage[['Regionai', 'Entity','Cellular Subscription', 'Internet Users(%)','Broadband Subscription']]

countries_avg = countries_by_region.groupby(['Regionai','Entity']).mean()

countries_avg
```

```python
# Atotrūkis tarp cellular ir broadband, visų šalių.

salys_cell_broad = internet_usage[['Entity', 'Cellular Subscription', 'Broadband Subscription']]

salys_cell_broad_avg = salys_cell_broad.groupby(['Entity'], as_index = False).mean()

salys_cell_broad_avg['Difference Cellular/Broadband'] = salys_cell_broad_avg['Cellular Subscription'] - salys_cell_broad_avg['Broadband Subscription']

salys_cell_broad_avg
```

```python
# Atfiltruojam tik Lietuvos skirtumą tarp cellular ir broadband.

lithuania_diff = salys_cell_broad_avg.loc[(salys_cell_broad_avg['Entity'] == 'Lithuania')]

lithuania_diff

```

```python
# YOY paskaičiavimas Internet Users growth.

yoy = internet_usage[['Year', 'Internet Users(%)']]

yoy_avg = yoy.groupby(['Year'], as_index = False).mean()

yoy_avg

```

```python
# Kai turime visų metų interneto vartotojų skaičių, paskaičiuojam  YOY growth su pct_change.

import pandas as pd

yoy_avg['Growth_Rate'] = yoy_avg['Internet Users(%)'].pct_change(periods=1) * 100
yoy_avg
```

```python
yoy_avg.plot(figsize=(8,5), kind = 'bar', x = 'Year', y = 'Growth_Rate', ylabel = 'Growth rate(%)', title = 'YOY growth rate')
```
